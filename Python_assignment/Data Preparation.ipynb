### Finding & Working on types of Breachs

**Main Objective :**
the data present in this column is composite & contains combinations of different classes, hence our main aim is to find these classes & separate them, which can be later on included as ONE HOT encodings in future.

```python
data['Type of Breach'].unique()
```

```python
types_breach = []

for i in data['Type of Breach'].unique():
    breach_type = (str(i).split(','))
    types_breach.append(breach_type)
```

```python
types_breach
```

```python
final_types = []
for i in types_breach:
    for j in i:
        final_types.append(j.strip())
        
final_types = set(final_types)
```

```python
final_breach_types = final_types
final_breach_types
```

**This is the final types of BREACH TYPES**

```python
len(final_types)
```

```python
breach_counts = dict.fromkeys(final_breach_types, 0) 
breach_counts
```

```python
# Getting the count for each individual types of breach
for breach in data['Type of Breach']:
    for breach_type in final_breach_types:
        if str(breach).find(breach_type) != -1:
            breach_counts[breach_type] += 1
```
